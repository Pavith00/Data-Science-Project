```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re

# Load the dataset (replace 'data' with your actual DataFrame variable if necessary)
# data = pd.read_csv('path_to_your_dataset.csv')

# Preprocess the text data: convert to lowercase, remove punctuation and stop words
def preprocess_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()
        # Remove punctuation and non-alphabetic characters
        text = re.sub(r'[^a-z\s]', '', text)
        return text
    else:
        return ''

data['Processed_Review_Text'] = data['Review_Text'].apply(preprocess_text)

# Tokenize and count word frequencies
word_counts = Counter()
for text in data['Processed_Review_Text']:
    words = text.split()
    filtered_words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    word_counts.update(filtered_words)

# Get the top 20 most frequent words
top_20_words = word_counts.most_common(20)

# Convert to DataFrame for plotting
top_20_df = pd.DataFrame(top_20_words, columns=['Word', 'Frequency'])

# Plot the top 20 words
plt.figure(figsize=(12, 8))
plt.bar(top_20_df['Word'], top_20_df['Frequency'], color='skyblue')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Top 20 Most Frequently Occurring Words')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# prompt: give me top 10 words that frequently occuring in reviews in rating 4 and 5

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import re

# Filter reviews with ratings 4 and 5
filtered_reviews = data[data['Rating'].isin([4, 5])]

# Preprocess the text data: convert to lowercase, remove punctuation and stop words
def preprocess_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()
        # Remove punctuation and non-alphabetic characters
        text = re.sub(r'[^a-z\s]', '', text)
        return text
    else:
        return ''

filtered_reviews['Processed_Review_Text'] = filtered_reviews['Review_Text'].apply(preprocess_text)

# Tokenize and count word frequencies
word_counts = Counter()
for text in filtered_reviews['Processed_Review_Text']:
    words = text.split()
    filtered_words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    word_counts.update(filtered_words)

# Get the top 10 most frequent words
top_10_words = word_counts.most_common(10)

# Print the top 10 words
print("Top 10 Most Frequently Occurring Words in Ratings 4 and 5:")
for word, count in top_10_words:
    print(f"- {word}: {count}")
```

```python
# prompt: give me most occuring words with next word in rating 1 2

# Filter reviews with ratings 1 and 2
filtered_reviews = data[data['Rating'].isin([1, 2])]

# Preprocess the text data: convert to lowercase, remove punctuation and stop words
def preprocess_text(text):
    if isinstance(text, str):
        # Convert to lowercase
        text = text.lower()
        # Remove punctuation and non-alphabetic characters
        text = re.sub(r'[^a-z\s]', '', text)
        return text
    else:
        return ''

filtered_reviews['Processed_Review_Text'] = filtered_reviews['Review_Text'].apply(preprocess_text)

# Tokenize and count word frequencies
word_counts = Counter()
for text in filtered_reviews['Processed_Review_Text']:
    words = text.split()
    filtered_words = [word for word in words if word not in ENGLISH_STOP_WORDS]
    word_counts.update(filtered_words)

# Get the top 10 most frequent words
top_10_words = word_counts.most_common(10)

# Print the top 10 words
print("Top 10 Most Frequently Occurring Words in Ratings 1 and 2:")
for word, count in top_10_words:
    print(f"- {word}: {count}")

# Count occurrences of next word after each top word
next_word_counts = {}
for word, _ in top_10_words:
    next_word_counts[word] = Counter()
    for text in filtered_reviews['Processed_Review_Text']:
        words = text.split()
        for i in range(len(words) - 1):
            if words[i] == word:
                next_word_counts[word][words[i + 1]] += 1

# Print the most frequent next word for each top word
for word, next_word_counts in next_word_counts.items():
    most_frequent_next_word, _ = next_word_counts.most_common(1)[0]
    print(f"Most frequent next word after '{word}': '{most_frequent_next_word}'")
```
